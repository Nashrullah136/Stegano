```python
import cv2
import os
import matplotlib.pyplot as plt
import math
```

```python
video = cv2.VideoCapture('./sample/video/ex2.avi')
```

```python
frames = []

while(True):
    ret, frame = video.read()
    
    if ret:
        frames.append(frame)
    else:
        break
```

```python
img = cv2.cvtColor(frames[0], cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
(major_ver, minor_ver, subminor_ver) = (cv2.__version__).split('.')

if int(major_ver) < 3:
    fps = video.get(cv2.cv.CV_CAP_PROP_FPS)
else:
    fps = video.get(cv2.CAP_PROP_FPS)
    
fps = int(math.ceil(fps))
print(fps)
```

```python
img_height, img_width, img_channels = frames[0].shape

size = (img_width, img_height)
output = cv2.VideoWriter('ex2_modified.avi', cv2.VideoWriter_fourcc(*'DIVX'), fps, size)

frames_len = len(frames)
for i in range(frames_len):
    output.write(frames[i])
    
output.release()
```
